### Lets get the required columns from mathermaticaly

```python
# categorical column finding using  chi square test
from scipy.stats import chi2_contingency
R_L = []
for i in cat_col.columns:
    
    data = pd.crosstab(df[i], df['Stress_Level'])
    # Chi-square test
    chi2, p, dof, expected = chi2_contingency(data)

    print(15*'-',i,15*'-')

    print("Chi-Square Statistic:", round(chi2, 2))
    print("Degrees of Freedom:", dof)
    print("p-value:", round(p, 4))
    print("Expected Frequencies:\n", expected)

    # Interpretation
    alpha = 0.05
    if p < alpha:
        print(f"Reject the null hypothesis â†’ {i} and Stress_Level are related.")
        R_L.append(i)
    else:
        print("Fail to reject null hypothesis â†’ No relationship.")
    print(15*'*',i,15*'*')
print(R_L)
cat_col = cat_col[R_L]
cat_col
```

```python
R_L = []
for i in num_dis.columns:
    
    data = pd.crosstab(df[i], df['Stress_Level'])
    # Chi-square test
    chi2, p, dof, expected = chi2_contingency(data)

    print(15*'-',i,15*'-')

    print("Chi-Square Statistic:", round(chi2, 2))
    print("Degrees of Freedom:", dof)
    print("p-value:", round(p, 4))
    print("Expected Frequencies:\n", expected)

    # Interpretation
    alpha = 0.05
    if p < alpha:
        print(f"Reject the null hypothesis â†’ {i} and Stress_Level are related.")
        R_L.append(i)
    else:
        print("Fail to reject null hypothesis â†’ No relationship.")
    print(15*'*',i,15*'*')
print(R_L)
num_dis = cat_col[R_L]
num_dis
```

```python
# Numeerical Continuos column selection using Anova Test due to having 3 classes
from scipy.stats import f_oneway

R_L = []

for i in num_con.columns:

    print(15*'-',i,15*'-')

    class_a = df[df['Stress_Level'] == 'Low'][i].to_list()
    class_b = df[df['Stress_Level'] == 'Medium'][i].to_list()
    class_c = df[df['Stress_Level'] == 'High'][i].to_list()

    # Perform one-way ANOVA
    f_stat, p_val = f_oneway(class_a, class_b, class_c)

    print(f"F-statistic: {f_stat:.4f}")
    print(f"p-value: {p_val:.4f}")

    # Interpret result
    alpha = 0.05
    if p_val < alpha:
        print("Reject the null hypothesis: At least one group is significantly different.")
        R_L.append(i)
    else:
        print("Fail to reject the null hypothesis: No significant difference.")
    
    print(15*'*',i,15*'*')
    print(' ')
print(R_L)
num_con = num_con[R_L]
num_con
```
